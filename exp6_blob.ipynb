# KNN classifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X,Y = make_blobs(n_samples=500,n_features=2,centers=4,cluster_std=1.5,random_state=4)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[:,0],X[:,1], c = Y, marker = '*',s = 100,edgecolors='black')
```

```python
X_tr, X_te, Y_tr, Y_te = train_test_split(X,Y,random_state=0)
```

```python
knn5 = KNeighborsClassifier(n_neighbors=5)
knn1 = KNeighborsClassifier(n_neighbors=1)
```

```python
knn5.fit(X_tr,Y_tr)
knn1.fit(X_tr,Y_tr)

y_pred5 = knn5.predict(X_te)
y_pred1 = knn1.predict(X_te)
```

```python
print("Accuracy with k=5: ", accuracy_score(Y_te, y_pred5)*100)
print("Accuracy with k=1: ", accuracy_score(Y_te, y_pred1)*100)
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.scatter(X_te[:,0], X_te[:,1], c=y_pred5, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=5", fontsize=20)
plt.subplot(1,2,2)
plt.scatter(X_te[:,0], X_te[:,1], c=y_pred1, marker= '*', s=100,edgecolors='black')
plt.title("Predicted values with k=1", fontsize=20)
plt.show()
```
