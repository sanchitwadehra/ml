```python
import pandas as pd
import statistics as st
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('exp2.csv')
```

```python
data.isnull().sum()
```

```python
print(data)
```

```python
corr_matrix_pearson=data.corr(method='pearson')
data.corr(method='pearson')
# Specifies the correlation method to use (default is Pearson):
# pearson: Standard correlation coefficient for linear relationships.
# kendall: Correlation coefficient for rank-based relationships.
# spearman: Another rank-based correlation coefficient, less sensitive to outliers.
```

```python
corr_matrix_kendall=data.corr(method='kendall')
data.corr(method='kendall')
```

```python
corr_matrix_spearman=data.corr(method='spearman')
data.corr(method='spearman')
```

```python
sns.heatmap(corr_matrix_pearson, cmap="coolwarm", annot=True)
```

```python
sns.heatmap(corr_matrix_kendall, cmap="coolwarm",annot=True)
```

```python
sns.heatmap(corr_matrix_spearman, cmap="coolwarm",annot=True)
```
